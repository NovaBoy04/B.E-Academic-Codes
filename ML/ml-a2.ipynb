```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, mean_absolute_error
```

```python
df = pd.read_csv('../input/email-spam-classification-dataset-csv/emails.csv')
df.head()
```

```python
df.describe()
```

```python
df.isnull().any()
```

```python
x = df.iloc[:,1:-1]
y = df.iloc[:,-1]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=35)
```

```python
knn = KNeighborsClassifier()
svm = SVC()
```

```python
knn.fit(x_train, y_train)
svm.fit(x_train, y_train)
```

```python
knn_pred = knn.predict(x_test)
svm_pred = svm.predict(x_test)
```

```python
print("Accuracy of KNN : ", accuracy_score(y_test,knn_pred)*100)
print("Mean Absolute Error: ", mean_absolute_error(y_test, knn_pred))
print("Recall: ", recall_score(y_test, knn_pred))
print("Precision: ", precision_score(y_test, knn_pred))
print("F1 Score: ", f1_score(y_test, knn_pred))

print("\nAccuracy of SVM : ", accuracy_score(y_test, svm_pred)*100)
print("Mean Absolute Error: ", mean_absolute_error(y_test, svm_pred))
print("Recall: ", recall_score(y_test, svm_pred))
print("Precision: ", precision_score(y_test, svm_pred))
print("F1 Score: ", f1_score(y_test, svm_pred))
```
